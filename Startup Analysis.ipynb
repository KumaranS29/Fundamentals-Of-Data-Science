```python
print("\nMeasures of Central Tendency:")
for column in df.select_dtypes(include=[np.number]).columns:
    print(f"\nColumn: {column}")
    print(f"Mean: {np.mean(df[column]):.2f}")
    print(f"Median: {np.median(df[column]):.2f}")
    print(f"Mode: {df[column].mode()[0]:.2f}")
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
if not categorical_columns.empty:
    print("\nMode for Categorical Columns:")
    for column in categorical_columns:
        print(f"{column}: {df[column].mode()[0]}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
numerical_columns = df.select_dtypes(include=[np.number]).columns
for column in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], kde=True, bins=20)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
nd =df.select_dtypes(include=[np.number])
correlation = nd.corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation,annot=True,cmap="coolwarm",fmt=".2f",linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
